```python
# Assuming 'df_19' is your DataFrame with columns 'state' and 'party'

# Group by 'state' and count the occurrences of each party
state_party_counts = df_19.groupby(['state', 'party']).size().reset_index(name='count')

# Find the party with the maximum count in each state
party_with_most_wins = state_party_counts.loc[state_party_counts.groupby('state')['count'].idxmax()]

party_with_most_wins
```
